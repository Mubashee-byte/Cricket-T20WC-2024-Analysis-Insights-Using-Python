**4Ô∏è‚É£ List All Unique Teams**

```python
import pandas as pd                                                            # Importing pandas library for data handling

df = pd.read_csv("T20WC.csv")                                                  # Read the dataset from the CSV file


teams = pd.unique(df[['1st Team', '2nd Team']].values.ravel('K'))              # Find all unique team names from both columns 'Team1' and 'Team2'

display(teams)
```

**5Ô∏è‚É£ Count Total Matches Played**

**6Ô∏è‚É£ Find Total Matches Won by Each Team**

**7Ô∏è‚É£ Find Matches Where Toss Winner Also Won the Match**

```python
import pandas as pd                                                            # Importing pandas library for data handling

df = pd.read_csv("T20WC.csv")                                                  # Read the dataset from the CSV file


same = df[df['Toss Winning'] == df['Winning Margin']]                              # Filter rows where toss winner and match winner are the same

display(same[['Match No.', 'Toss Winning', 'Winning Margin']])                        # Display selected details for clarity
```

**8Ô∏è‚É£ List Top 5 Highest First Innings Scores**

```python
import pandas as pd                                                            # Importing pandas library for data handling

df = pd.read_csv("T20WC.csv")                                                  # Read the dataset from the CSV file

top_scores = df.nlargest('First Innings Score')[['Match No.', '1st Team', '2nd Team', 'First Innings Score']]          # Find the top 5 matches with the highest first innings score

display(top_scores)
```

**9Ô∏è‚É£ Count Matches by Stage**

**üîü Find Average Winning Margin**


```python
import pandas as pd                                                            # Importing pandas library for data handling
import numpy as np

df = pd.read_csv("T20WC.csv")                                                  # Read the dataset from the CSV file

# Replace 'Rain' with NaN and convert the column to numeric
df['Winning Margin'] = df['Winning Margin'].replace('Rain', np.nan).astype(float)

avg_margin = df['Winning Margin'].mean()                                       # Calculate the average winning margin across all matches


display("Average Winning Margin:", avg_margin)                         # Round the result for neat display
```

**1Ô∏è‚É£2Ô∏è‚É£ Show Top 5 Players with Most ‚ÄúPlayer of the Match‚Äù Awards**


```python
import pandas as pd                                                            # Importing pandas library for data handling

df = pd.read_csv("T20WC.csv")                                                  # Read the dataset from the CSV file

top_players = df['Player Of The Match'].value_counts().head(5)                  # Count how many times each player won the Player of the Match award

display(top_players)
```

**1Ô∏è‚É£3Ô∏è‚É£ Count Matches Abandoned Due to Rain**


```python
import pandas as pd                                                                     # Importing pandas library for data handling

df = pd.read_csv("T20WC.csv")                                                          # Read the dataset from the CSV file


abandoned = df[df['Winners'].str.contains('abandoned|rain', case=False, na=False)] # Filter matches that were abandoned or affected by rain

display("Matches Abandoned due to Rain:", len(abandoned))                                  # Count how many such matches occurred
```

**1Ô∏è‚É£4Ô∏è‚É£ Find Venue Hosting Most Matches**
